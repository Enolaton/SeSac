```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_all = pd.read_csv('all_category_summary_202201_202511.csv')
print(data_all.shape)
print(data_all.info())
print(data_all.isnull().sum())
print(data_all.head())
```

```python
df_f_20s = data_all[(data_all['sex'] == 'F') & (data_all['age_group'] == '30대')]

cnt_by_food_f_30s = df_f_20s.groupby('card_tpbuz_nm_2')['cnt'].sum()

cnt_by_food_f_30s.sort_values(ascending=False)
```

```python
df_f_20s = data_all[(data_all['sex'] == 'F') & (data_all['age_group'] == '20대')]

cnt_by_food_f_20s = df_f_20s.groupby('card_tpbuz_nm_2')['cnt'].sum()

cnt_by_food_f_20s.sort_values(ascending=False)
```

```python
plt.figure(figsize=(7, 7))

cnt_by_food_f_20s.plot.pie(autopct='%1.1f%%')

plt.title("20대 여자의 카테고리별 카드 결제 횟수")

plt.ylabel('')

plt.show()
```

```python
# 20대 남자
df_m_20s = data_all[(data_all['sex'] == 'M') & (data_all['age_group'] == '20대')]
cnt_by_food_m_20s = df_m_20s.groupby('card_tpbuz_nm_2')['cnt'].sum()
cnt_by_food_m_20s.sort_values(ascending=False)

# 30대 남자
df_m_30s = data_all[(data_all['sex'] == 'M') & (data_all['age_group'] == '30대')]
cnt_by_food_m_30s = df_m_30s.groupby('card_tpbuz_nm_2')['cnt'].sum()
cnt_by_food_m_30s.sort_values(ascending=False)

# 40대 남자
df_m_40s = data_all[(data_all['sex'] == 'M') & (data_all['age_group'] == '40대')]
cnt_by_food_m_40s = df_m_40s.groupby('card_tpbuz_nm_2')['cnt'].sum()
cnt_by_food_m_40s.sort_values(ascending=False)

# 50대 남자
df_m_50s = data_all[(data_all['sex'] == 'M') & (data_all['age_group'] == '50대')]
cnt_by_food_m_50s = df_m_50s.groupby('card_tpbuz_nm_2')['cnt'].sum()
cnt_by_food_m_50s.sort_values(ascending=False)


# 60대이상 남자
df_m_60s = data_all[(data_all['sex'] == 'M') & (data_all['age_group'] == '60대 이상')]
cnt_by_food_m_60s = df_m_60s.groupby('card_tpbuz_nm_2')['cnt'].sum()
cnt_by_food_m_60s.sort_values(ascending=False)

# 20대 여자
df_f_20s = data_all[(data_all['sex'] == 'F') & (data_all['age_group'] == '20대')]
cnt_by_food_f_20s = df_f_20s.groupby('card_tpbuz_nm_2')['cnt'].sum()
cnt_by_food_f_20s.sort_values(ascending=False)

# 30대 여자
df_f_30s = data_all[(data_all['sex'] == 'F') & (data_all['age_group'] == '30대')]
cnt_by_food_f_30s = df_f_30s.groupby('card_tpbuz_nm_2')['cnt'].sum()
cnt_by_food_f_30s.sort_values(ascending=False)

# 40대 여자
df_f_40s = data_all[(data_all['sex'] == 'F') & (data_all['age_group'] == '40대')]
cnt_by_food_f_40s = df_f_40s.groupby('card_tpbuz_nm_2')['cnt'].sum()
cnt_by_food_f_40s.sort_values(ascending=False)

# 50대 여자
df_f_50s = data_all[(data_all['sex'] == 'F') & (data_all['age_group'] == '50대')]
cnt_by_food_f_50s = df_f_50s.groupby('card_tpbuz_nm_2')['cnt'].sum()
cnt_by_food_f_50s.sort_values(ascending=False)


# 60대이상 여자
df_f_60s = data_all[(data_all['sex'] == 'F') & (data_all['age_group'] == '60대 이상')]
cnt_by_food_f_60s = df_f_60s.groupby('card_tpbuz_nm_2')['cnt'].sum()
cnt_by_food_f_60s.sort_values(ascending=False)
```

```python
fig, axes = plt.subplots(2, 5, figsize=(25, 12))

axes[0,0].pie(cnt_by_food_m_20s, labels=cnt_by_food_m_20s.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
axes[0,0].set_title("20대 남자")

axes[0,1].pie(cnt_by_food_m_30s, labels=cnt_by_food_m_20s.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
axes[0,1].set_title("30대 남자")

axes[0,2].pie(cnt_by_food_m_40s, labels=cnt_by_food_m_20s.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
axes[0,2].set_title("40대 남자")

axes[0,3].pie(cnt_by_food_m_50s, labels=cnt_by_food_m_20s.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
axes[0,3].set_title("50대 남자")

axes[0,4].pie(cnt_by_food_m_60s, labels=cnt_by_food_m_20s.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
axes[0,4].set_title("60대이상 남자")

axes[1,0].pie(cnt_by_food_f_20s, labels=cnt_by_food_m_20s.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
axes[1,0].set_title("20대 여자")

axes[1,1].pie(cnt_by_food_f_30s, labels=cnt_by_food_m_20s.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
axes[1,1].set_title("30대 여자")

axes[1,2].pie(cnt_by_food_f_40s, labels=cnt_by_food_m_20s.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
axes[1,2].set_title("40대 여자")

axes[1,3].pie(cnt_by_food_f_50s, labels=cnt_by_food_m_20s.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
axes[1,3].set_title("50대 여자")

axes[1,4].pie(cnt_by_food_f_60s, labels=cnt_by_food_m_20s.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
axes[1,4].set_title("60대이상 여자")
```
