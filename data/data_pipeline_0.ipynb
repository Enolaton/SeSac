```python
import polars as pl
import glob

# 1. 모든 CSV 파일 경로 가져오기
file_paths = glob.glob('../food/*.csv') 

# 2. Polars를 이용해 고속 변환 (메모리를 아끼기 위해 스캐닝 방식 사용)
for path in file_paths:
    output_name = path.replace('.csv', '.parquet')
    pl.scan_csv(path).sink_parquet(output_name)
    print(f"Converted: {output_name}")
```

```python
# Parquet 파일들을 스캔 (실제 로드가 아닌 연결만 함)
queries = pl.scan_parquet("../food/*.parquet")

# 최신 1년치 필터링 및 특정 지역(예: 강남구) 샘플링
# ta_ymd가 정수형인 경우 20241201 이상인 데이터만 추출
recent_year_data = queries.filter(
    pl.col("ta_ymd") >= 20241201
).collect() # 여기서 실제 메모리로 로드

print(recent_year_data.shape)
```
