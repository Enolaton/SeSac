======================= Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ ÏÜåÎπÑ, Í≤∞Ï†ú ÌöüÏàò ÌôïÏù∏ =======================

```python
import pandas as pd
import glob

DATA_PATH = "../food/*.csv"
file_list = sorted(glob.glob(DATA_PATH))

amt_sum = pd.Series(dtype="float64")
cnt_sum = pd.Series(dtype="float64")

for file in file_list:
    print(f"Processing {file}")

    for chunk in pd.read_csv(
        file,
        usecols=["card_tpbuz_nm_2", "amt", "cnt"],
        chunksize=200_000
    ):
        chunk["card_tpbuz_nm_2"] = chunk["card_tpbuz_nm_2"].astype("category")

        grouped = chunk.groupby("card_tpbuz_nm_2").agg(
            amt=("amt", "sum"),
            cnt=("cnt", "sum")
        )

        amt_sum = amt_sum.add(grouped["amt"], fill_value=0)
        cnt_sum = cnt_sum.add(grouped["cnt"], fill_value=0)

df_overall = (
    pd.DataFrame({
        "amt": amt_sum,
        "cnt": cnt_sum
    })
    .sort_values("amt", ascending=False)
)
```

```python
# ÎßåÎì§Ïñ¥ÏßÑ df_overallÏùÑ csvÎ°ú Ï†ÄÏû•ÌïòÍ∏∞
import os

OUTPUT_DIR = "../output"

df_overall_out = df_overall.reset_index()

df_overall_out.columns = ["category", "total_amt", "total_cnt"]

df_overall_out.to_csv(
    f"{OUTPUT_DIR}/overall_category_summary_202201_202511.csv",
    index=False,
    encoding="utf-8-sig"
)
```

======================= ÏõîÎ≥Ñ ÏÜåÎπÑ, Í≤∞Ï†ú ÌöüÏàò Îì± ÌôïÏù∏ =======================

```python
import pandas as pd
import glob
import os

DATA_PATH = "../food/*.csv"
file_list = sorted(glob.glob(DATA_PATH))

monthly_data = []

for file in file_list:
    year_month = os.path.basename(file).replace(".csv", "")

    # ÌååÏùº Îã®ÏúÑ ÎàÑÏ†ÅÏö© Series
    amt_sum = pd.Series(dtype="float64")
    cnt_sum = pd.Series(dtype="float64")

    for chunk in pd.read_csv(
        file,
        usecols=["card_tpbuz_nm_2", "amt", "cnt"],
        chunksize=200_000
    ):
        # category Î©îÎ™®Î¶¨ ÏµúÏ†ÅÌôî
        chunk["card_tpbuz_nm_2"] = chunk["card_tpbuz_nm_2"].astype("category")

        grouped = chunk.groupby("card_tpbuz_nm_2").agg(
            amt=("amt", "sum"),
            cnt=("cnt", "sum")
        )

        amt_sum = amt_sum.add(grouped["amt"], fill_value=0)
        cnt_sum = cnt_sum.add(grouped["cnt"], fill_value=0)

    # ÌååÏùº ÌïòÎÇò ÏôÑÎ£å ‚Üí DataFrame ÏÉùÏÑ±
    if not amt_sum.empty:
        file_df = pd.DataFrame({
            "category": amt_sum.index,
            "amt": amt_sum.values,
            "cnt": cnt_sum.reindex(amt_sum.index, fill_value=0).values,
            "year_month": year_month
        })
        monthly_data.append(file_df)

# Ï†ÑÏ≤¥ Î≥ëÌï© (1Î≤àÎßå)
df_monthly = pd.concat(monthly_data, ignore_index=True)
```

```python
import os

OUTPUT_DIR = "../output"
os.makedirs(OUTPUT_DIR, exist_ok=True)

df_monthly_out = df_monthly.copy()

df_monthly_out.to_csv(
    f"{OUTPUT_DIR}/monthly_category_summary_202201_202511.csv",
    index=False,
    encoding="utf-8-sig"
)
```
