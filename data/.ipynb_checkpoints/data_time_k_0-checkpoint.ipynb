```python
import pandas as pd
import glob
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

# 한글 폰트 설정
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

# 1. 설정 및 매핑 데이터
DATA_PATH = "../food"
DAY_MAP = {1: '월', 2: '화', 3: '수', 4: '목', 5: '금', 6: '토', 7: '일'}
HOUR_MAP = {
    1: '00:00~06:59', 2: '07:00~08:59', 3: '09:00~10:59', 4: '11:00~12:59',
    5: '13:00~14:59', 6: '15:00~16:59', 7: '17:00~18:59', 8: '19:00~20:59',
    9: '21:00~22:59', 10: '23:00~23:59'
}
day_order = ['월', '화', '수', '목', '금', '토', '일']
hour_order = [HOUR_MAP[i] for i in range(1, 11)]

# 2. 파일 목록 확인
files = sorted(glob.glob(os.path.join(DATA_PATH, "*.csv")))
if not files:
    raise ValueError("CSV 파일을 찾지 못했습니다. 경로를 확인하세요.")

final_result = []

# 3. 데이터 로드 및 집계 (Chunking)
print(f"총 {len(files)}개의 파일 처리를 시작합니다...")

for file in files:
    print(f"현재 파일 처리 중: {os.path.basename(file)}")
    
    # 필요한 컬럼만 로드하여 메모리 절약
    reader = pd.read_csv(file, usecols=["hour", "day", "amt"], chunksize=500_000)
    
    for chunk in reader:
        # 매핑 적용
        chunk["day_str"] = chunk["day"].map(DAY_MAP)
        chunk["hour_str"] = chunk["hour"].map(HOUR_MAP)
        
        # 즉시 그룹화하여 데이터 크기 축소
        agg = chunk.groupby(["hour_str", "day_str"])["amt"].sum().reset_index()
        final_result.append(agg)

# 4. 최종 데이터 병합
df_final = pd.concat(final_result, ignore_index=True)
df_final = df_final.groupby(["hour_str", "day_str"])["amt"].sum().reset_index()

# 5. 피벗 테이블 생성 및 정렬
pivot_df = df_final.pivot(index='hour_str', columns='day_str', values='amt')
pivot_df = pivot_df.reindex(index=hour_order, columns=day_order)

# 6. 시각화 (Log Scale 적용)
plt.figure(figsize=(15, 8))
# 데이터 편차가 클 경우 패턴 확인을 위해 log1p 사용
sns.heatmap(np.log1p(pivot_df), 
            cmap='Reds', 
            annot=False, 
            linewidths=0.1,
            cbar_kws={'label': '소비 금액 (Log Scale)'})

plt.title('전체 요일 × 시간대 소비 패턴', fontsize=18, pad=20)
plt.xlabel('요일', fontsize=12)
plt.ylabel('시간대', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import glob
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

# 1. 환경 설정
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

DATA_PATH = "../food"
DAY_MAP = {1: '월', 2: '화', 3: '수', 4: '목', 5: '금', 6: '토', 7: '일'}
HOUR_MAP = {
    1: '00:00~06:59', 2: '07:00~08:59', 3: '09:00~10:59', 4: '11:00~12:59',
    5: '13:00~14:59', 6: '15:00~16:59', 7: '17:00~18:59', 8: '19:00~20:59',
    9: '21:00~22:59', 10: '23:00~23:59'
}
DAY_ORDER = ['월', '화', '수', '목', '금', '토', '일']
HOUR_ORDER = [HOUR_MAP[i] for i in range(1, 11)]

# 2. 파일 로드 및 카테고리별 집계
files = sorted(glob.glob(os.path.join(DATA_PATH, "*.csv")))
final_result = []

print(f"데이터 읽기 시작 (총 {len(files)}개 파일)...")

for file in files:
    # 'card_tpbuz_nm_2' 컬럼을 반드시 포함하여 읽습니다.
    reader = pd.read_csv(file, usecols=["hour", "day", "card_tpbuz_nm_2", "amt"], chunksize=500_000)
    
    for chunk in reader:
        # 데이터 타입 정제 (이전 카테고리 정보가 남지 않도록 str 변환)
        chunk["card_tpbuz_nm_2"] = chunk["card_tpbuz_nm_2"].astype(str).str.strip()
        
        # 매핑 적용
        chunk["day_str"] = chunk["day"].map(DAY_MAP)
        chunk["hour_str"] = chunk["hour"].map(HOUR_MAP)
        
        # 업종별/시간별/요일별 합계 계산
        agg = chunk.groupby(["card_tpbuz_nm_2", "hour_str", "day_str"])["amt"].sum().reset_index()
        final_result.append(agg)

# 데이터 통합
df_final = pd.concat(final_result, ignore_index=True)
df_final = df_final.groupby(["card_tpbuz_nm_2", "hour_str", "day_str"])["amt"].sum().reset_index()

# 4. 히트맵 그리기 함수
def draw_category_heatmap(df, category_name):
    # 해당 카테고리만 필터링
    sub_df = df[df["card_tpbuz_nm_2"] == category_name]
    
    # 피벗 테이블 생성
    pivot = sub_df.pivot_table(index="hour_str", columns="day_str", values="amt", aggfunc="sum")
    pivot = pivot.reindex(index=HOUR_ORDER, columns=DAY_ORDER).fillna(0)
    
    # 시각화
    plt.figure(figsize=(12, 6))
    sns.heatmap(np.log1p(pivot), cmap="Reds", annot=False, linewidths=.5)
    
    plt.title(f"[{category_name}] 요일 × 시간대 소비 패턴 (Log Scale)", fontsize=15)
    plt.xlabel("요일")
    plt.ylabel("시간대")
    plt.tight_layout()
    plt.show()

# 5. 모든 카테고리 순회하며 그리기
categories = sorted(df_final["card_tpbuz_nm_2"].unique())
for cat in categories:
    draw_category_heatmap(df_final, cat)
```
